```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    'Date': pd.date_range(start='2023-01-01', periods=12, freq='M'),
    'Bitcoin': [5, -3, 10, 7, -4, 12, 8, -2, 6, 9, -1, 4],
    'Ethereum': [4, -2, 8, 6, -3, 10, 7, -1, 5, 8, 0, 3],
    'Stock_Index': [2, 1, 3, 2, -1, 4, 3, 1, 2, 3, 0, 1]
}

df = pd.DataFrame(data)
df
```

```python
df.describe()
```

Este dataset cont√©m retornos mensais simulados de ativos financeiros.

```python
returns = (1 + df[['Bitcoin', 'Ethereum', 'Stock_Index']] / 100).cumprod()
returns
```

```python
returns.plot(figsize=(10,5))
plt.title('Retorno Acumulado dos Ativos')
plt.ylabel('Crescimento do Investimento')
plt.xlabel('Per√≠odo')
plt.show()
```
